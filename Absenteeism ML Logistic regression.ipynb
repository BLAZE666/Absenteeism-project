```python
import pandas as pd
import numpy as np
```

```python
data_preprocessed = pd.read_csv('Absenteeism preprocessed.csv')
```

```python
data_preprocessed.head()
```

## Create the targets

```python
data_preprocessed['Absenteeism Time in Hours'].median()
```

if a person is absent for less then 2 hours then we will assign the value 0 (moderately absent) otherwise 1 (excessively absent)

```python
targets = np.where(data_preprocessed['Absenteeism Time in Hours'] > data_preprocessed['Absenteeism Time in Hours'].median(), 1, 0)
```

```python
targets
```

```python
data_preprocessed['Excessive Absenteeism'] = targets
```

```python
data_preprocessed
```

```python
targets.sum() / targets.shape[0]
```

```python
data_preprocessed['Absenteeism Time in Hours'].unique()
```

```python
data_preprocessed['Absenteeism Time in Hours'].value_counts()
```

```python
data_with_targets = data_preprocessed.drop(['Absenteeism Time in Hours'], axis =1)
```

```python
data_with_targets
```

```python
unscaled_inputs = data_with_targets.iloc[:,:-1]
```

```python
unscaled_inputs
```

```python
unscaled_inputs.columns.values
```

```python
columns_to_scale = ['Month value', 'Day of the week', 'Transportation Expense', 'Distance to Work',
       'Age', 'Daily Work Load Average', 'Body Mass Index',
       'Children', 'Pets']
```

```python
columns_not_scale = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Education']
```

```python
df_scaled = unscaled_inputs[columns_to_scale]
df_not_scaled = unscaled_inputs[columns_not_scale]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
absenteesim_scaler = StandardScaler()

absenteesim_scaler.fit(df_scaled)

scaled_inputs=absenteesim_scaler.transform(df_scaled)
```

```python
df_scaled = pd.DataFrame(scaled_inputs, columns=columns_to_scale)
scaled_inputs = pd.concat([df_scaled, df_not_scaled], axis=1)
scaled_inputs
```

```python
scaled_inputs.shape
```

```python
reordered_columns = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month value',
       'Day of the week', 'Transportation Expense', 'Distance to Work',
       'Age', 'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets']
```

```python
scaled_inputs = scaled_inputs[reordered_columns]
```

```python
scaled_inputs
```

## Spliting the data into test and train

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, train_size = 0.8, random_state=20)
```

```python
print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape, y_test.shape)
```

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg2 = LogisticRegression()
```

```python
reg2.fit(x_train, y_train)
```

```python
reg2.score(x_train, y_train)
```

### Manually check the accuracy

```python
model_output = reg2.predict(x_train)
model_output
```

```python
y_train
```

```python
model_output == y_train
```

```python
np.sum(model_output==y_train)/model_output.shape[0]
```

#### Finding the intercept and coeff

```python
reg2.intercept_
```

```python
reg2.coef_
```

```python
feature_names = unscaled_inputs.columns.values
```

```python
feature_names
```

```python
summary_table = pd.DataFrame(columns=['Feature names'], data=feature_names)
summary_table['Coefficient'] = np.transpose(reg2.coef_)
summary_table
```

```python
summary_table.index = summary_table.index + 1
summary_table.loc[0] = ['Intercept', reg2.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

```python
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficient)
```

```python
summary_table
```

```python
summary_table.sort_values('Odds_ratio', ascending=False)
```

From the summary table we can see the distance to work variable is very insignificant as the coeff value is close to zero and odds ratio is close to 1 so we can safely ommit the variable or just keep it as it has no significant impact on the model

## Testing the model

```python
reg2.score(x_test, y_test)
```

Test accuracy score achived 57%
